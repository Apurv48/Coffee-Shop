```python
rota_info = []
Rota_id = input("Enter rota id")
for i in range(0,len(rota)):
    if(rota.iloc[i,1]==Rota_id):
        for j in range(0,len(shift)):
            for k in range(0, len(staff)):
                if(rota.iloc[i,4]==staff.iloc[k,0]):
                     if(rota.iloc[i,3]==shift.iloc[j,0]):
                            rota_info.append([rota.iloc[i,1],rota.iloc[i,2],rota.iloc[i,3],rota.iloc[i,4],shift.iloc[j,1],shift.iloc[j,2],shift.iloc[j,3],staff.iloc[k,1],staff.iloc[k,2],staff.iloc[k,3],staff.iloc[k,4]])
rota_info = pd.DataFrame(rota_info)
rota_info = rota_info.rename(columns={0:'rota_id',1:"date",2:"shift_id",3:"staff_id",4:"day_of_week",5:"start_time",6:"end_time",7:"first_name",8:"last_name",9:"position",10:"sal_per_hour"})
rota_info
      
```

```python
date_info = []
Date = input("Enter Date")
for i in range(0,len(rota)):
    if(rota.iloc[i,2]==Date):
        for j in range(0,len(shift)):
            for k in range(0, len(staff)):
                if(rota.iloc[i,4]==staff.iloc[k,0]):
                     if(rota.iloc[i,3]==shift.iloc[j,0]):
                            date_info.append([rota.iloc[i,1],rota.iloc[i,2],rota.iloc[i,3],rota.iloc[i,4],shift.iloc[j,1],shift.iloc[j,2],shift.iloc[j,3],staff.iloc[k,1],staff.iloc[k,2],staff.iloc[k,3],staff.iloc[k,4]])
date_info = pd.DataFrame(date_info)
date_info = date_info.rename(columns={0:'rota_id',1:"date",2:"shift_id",3:"staff_id",4:"day_of_week",5:"start_time",6:"end_time",7:"first_name",8:"last_name",9:"position",10:"sal_per_hour"})
date_info
      
```

```python
shift_info = []
Shift_id = input("Enter Shift id")
for i in range(0,len(rota)):
    if(rota.iloc[i,3]==Shift_id):
        for j in range(0,len(shift)):
            for k in range(0, len(staff)):
                if(rota.iloc[i,4]==staff.iloc[k,0]):
                     if(rota.iloc[i,3]==shift.iloc[j,0]):
                            shift_info.append([rota.iloc[i,1],rota.iloc[i,2],rota.iloc[i,3],rota.iloc[i,4],shift.iloc[j,1],shift.iloc[j,2],shift.iloc[j,3],staff.iloc[k,1],staff.iloc[k,2],staff.iloc[k,3],staff.iloc[k,4]])
shift_info = pd.DataFrame(shift_info)
shift_info = shift_info.rename(columns={0:'rota_id',1:"date",2:"shift_id",3:"staff_id",4:"day_of_week",5:"start_time",6:"end_time",7:"first_name",8:"last_name",9:"position",10:"sal_per_hour"})
shift_info
      
```
